```python
def barPerc(df, xVar, ax):
    """
    source: https://stackoverflow.com/a/67076347/4852724
    
    barPerc(): Add percentage for hues to bar plots
    args:
        df: pandas dataframe
        xVar: (string) X variable
        ax: Axes object (for Seaborn Countplot/Bar plot or
                             pandas bar plot)
    """
    # 1. How many X categories
    ## check for NaN and remove
    numX = len([x for x in df[xVar].unique() if x==x])
    
    # 2. The bars are created in hue order, organize them
    bars = ax.patches
    ## 2a. For each X variable
    for ind in range(numX):
        ## 2b. Get hue bar
        ##     ex. 8 X categories, 4 hues =>
        ##     [0, 8, 16, 24] are hue bars for 1st X category
        hueBars = bars[ind:][::numX]
        ## 2c. Get the total height (for percentages)
        total = sum([x.get_height() for x in hueBars])
        
        # 3. Print the percentage on the bars
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,
                   bar.get_height() - 0.1 * bar.get_height(),
                   f"{bar.get_height()/total:.2%}",
                   ha='center', va='top', color='#f6f6f6')


def GrpSubplots(df, xVar, ax, axis=None, ticklabels=None):
    """
    Create subplots based on X variables
    """
    df.groupby([xVar, 'No-show']).size().unstack('No-show').apply(lambda x: x*100/x.sum(), axis=axis).plot.bar(rot=0,
                                                                                                           width=0.9,
                                                                                                           color=['#0e4f66', 'grey'],
                                                                                                           ax=ax);
    ax.set_title(f'{xVar}'.title(), fontsize=14, fontweight='bold', fontfamily='serif', color='#323232', loc='left')
    ax.grid(color='gray', linestyle=':', axis='y', zorder=0, dashes=(1,5))
    ax.set_xticklabels(ticklabels)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    ax.yaxis.set_major_locator(mtick.MultipleLocator(10))
    ax.get_legend().remove()
    
    barPerc(df, xVar, ax)
```

```python
background_color = '#fbfbfb'

fig = plt.figure(figsize=(20,10), dpi=150)
fig.patch.set_facecolor(background_color)
gs = fig.add_gridspec(2,3)
gs.update(wspace=0.35, hspace=0.27)

# axes as a list
axes = [fig.add_subplot(gs[0,0]),
        fig.add_subplot(gs[0,1]),
        fig.add_subplot(gs[0,2]),
        fig.add_subplot(gs[1,0]),
        fig.add_subplot(gs[1,1]),
        fig.add_subplot(gs[1,2])]

# categories as a list
cats = ['Gender', 'Scholarship', 'Hipertension', 'Diabetes', 'SMS_received', 'Alcoholism']

# ticklabels as a list
tlabs = [['Female', 'Male']] + [['No', 'Yes']]*5

axes[0].text(-0.80, 145, 'Show', fontsize=20, 
        fontweight='bold', fontfamily='serif', color='#0e4f66')
axes[0].text(-0.05, 145, 'vs.', fontsize=20,
        fontweight='bold', fontfamily='serif', color='#323232')
axes[0].text(0.35, 145, 'No-Show', fontsize=20,
        fontweight='bold', fontfamily='serif', color='gray')
axes[0].text(-0.80, 136, '- The proportions of Gender, Diabetes, Alcoholism on No-show and Show groups look pretty similar to each other.',
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.80, 130, '- Hipertension and Scholarship might have interaction between No-show and Show groups.',
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.80, 124, '- SMS_received looked like there would be interaction between No-show and SMS_received.',
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.70, 118, '- For the no-show group, when they did not get SMS reminders, they are less likely to show on the schedule dates.',
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.70, 112, "- For the show group, people showed on the schedule dates, even though they didn't get SMS.",
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.70, 106, "- The show group are less likely to show on the schedule dates when they didn't get SMS compared to the no-show group.",
        fontsize=14, fontfamily='serif', color='#323232')
axes[0].text(-0.70, 100, "- The no-show group are less likely to show on the schedule dates even though they got SMS compared to the show group.",
        fontsize=14, fontfamily='serif', color='#323232')

for ax, xVar, ticklabels in zip(axes, cats, tlabs):
    GrpSubplots(df, xVar, ax, axis=1, ticklabels=ticklabels)
```
